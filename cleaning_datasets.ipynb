### Gestion des valeurs manquantes

La valeur manquante dans le jeu de donnée peut autant signifier qu'il n'y a **pas eu de morts** dans le cas de la colonne **Total Deaths** ou que la donnée soit vraiment **manquante**.

```python
# Regardons le nombre de valeurs manquantes de la colonne Total Deaths par type de catastrophe naturelle
print(disasters.groupby('Disaster_type')['Total_deaths'].apply(
    lambda x: x.isnull().sum()
).sort_values(ascending=False))
```

On remarque que les valeurs manquantes sont concentrées sur les 3 catastrophes naturelles: **inondation, tempête, sécheresse**. Ce qui est logique étant donné que ce sont les catastrophes naturelles **les moins mortelles** pour la population. Donc, les valeurs manquantes ici représentent réellement une donnée. Pour les comptabiliser, on va les remplacer par 0.

```python
# Remplir les valeurs manquantes par 0
disasters['Total_deaths'] = disasters['Total_deaths'].fillna(0)
disasters['Total_affected'] = disasters['Total_affected'].fillna(0)

# Imputation par la médiane par type de catastrophe
columns = ['Total_deaths', 'Total_affected']

for col in columns:
    disasters[col] = disasters.groupby('Country')[col].transform(
        lambda x: x.fillna(x.median())
    )
    disasters[col] = disasters[col].fillna(disasters[col].median())
```

```python
# Aggrégation par année et par pays
disasters_agg = disasters.groupby(['Country', 'Year']).agg(
    nb_disasters     = ('Disaster_type', 'count'),
    total_deaths_per_country = ('Total_deaths', 'sum'),
    total_affected   = ('Total_affected', 'sum')
).reset_index()
```

### Simplification de l'âge

Voici les catégories d'âge disponsible :
- 10-19 years,
- 5-9 years,
- 10-14 years,
- 15-19 years,
- 15-24 years,
- 25-34 years,
- 35-44 years,
- 45-54 years,
- 55-64 years,
- 65-74 years,
- 75-84 years,
- 85+ years,
- 0-4 years,
- 18-29 years,
- 30-39 years,
- 40-49 years,
- 5-17 years,
- 50-59 years,
- 60-69 years,
- 70-79 years,
- et enfin 80+ years.

Nous allons les simplifier en 3 catégories principales :
- les -25 ans
- les -60 ans
- et les +60 ans

```python
def age_group(x):
    if pd.isna(x):
        return np.nan

    if x in [
        "0-4 years",
        "5-9 years",
        "10-14 years",
        "5-17 years",
        "15-19 years",
        "10-19 years",
        "15-19 years",
        "15-24 years"
    ]:
        return "-25"

    elif x in [
        "18-29 years",
        "25-34 years",
        "30-39 years",
        "35-44 years",
        "45-54 years",
        "50-59 years"
    ]:
        return "-60"

    elif x in [
        "55-64 years",
        "60-69 years",
        "65-74 years",
        "70-79 years",
        "75-84 years",
        "80+ years",
        "85+ years",
        "Only older adults (>=60 years)"
    ]:
        return "60+"

    return np.nan

health_inequality["age"] = health_inequality["group"].apply(age_group)
```

### Simplification des données sur le cholesterole

Voici les données disponibles sur le cholestérole :
- Cholesterol, mean high-density lipoprotein (HDL) (age-standardized),
- Cholesterol, mean high-density lipoprotein (HDL) (crude),
- Cholesterol, mean non-high-density lipoprotein (non-HDL) (age-standardized),
- Cholesterol, mean non-high-density lipoprotein (non-HDL) (crude),
- Cholesterol, mean total (age-standardized),
et enfi, Cholesterol, mean total (crude).

Ici nous avons fait le choix de garder uniquement les mesures standardisées en fonction de l'âge et du pays.
Ce sont les mesures les plus fiables à utiliser lors de comparaison internationales.

On va donc garder :
- Cholesterol, mean high-density lipoprotein (HDL) (age-standardized),
- Cholesterol, mean non-high-density lipoprotein (non-HDL) (age-standardized),
- et Cholesterol, mean total (age-standardized)

Qui seront respectivement renommé :
- good
- bad
- total


```python
def cholesterol_simplification(x):
    if pd.isna(x):
        return np.nan

    if x in ["Cholesterol, mean high-density lipoprotein (HDL) (age-standardized)"]:
        return "good"
    
    if x in ["Cholesterol, mean non-high-density lipoprotein (non-HDL) (age-standardized)"]:
        return "bad"
    
    if x in ["Cholesterol, mean total (age-standardized)"]:
        return "total"
    
health_inequality["cholesterol"] = health_inequality["indicator_name"].apply(cholesterol_simplification)
```

Désormais, attaquons nous à la simplification de la distance pour obtenir une aide technique (assistive techology)

### Simplification des annotations de la distance pour obtenir une aide technique (assistive techology)

Pour "assistive technology", nous avons les catégories suivantes :
- Assistive technology: Travel distance to obtain assistive products (%) - 100+ km,
- Assistive technology: Travel distance to obtain assistive products (%) - 26-50 km,
- Assistive technology: Travel distance to obtain assistive products (%) - 51-100 km,
- Assistive technology: Travel distance to obtain assistive products (%) - 6-25 km,
- Assistive technology: Travel distance to obtain assistive products (%) - <6 km,
- Assistive technology: Met need of assistive products (%),
- Assistive technology: Need of assistive products (%),
- et enfin, Assistive technology: Use of assistive products (%)

Nous allons le renommer par les 6 catégories suivantes :
- near
- close
- mid_distance
- need
- use
- et enfin, coverage

```python
def assistive_techno(x):
    if pd.isna(x):
        return np.nan
    
    if x in ["Assistive technology: Travel distance to obtain assistive products (%) - 100+ km",
             "Assistive technology: Travel distance to obtain assistive products (%) - 51-100 km"
            ]:
        return "near"
    
    if x in ["Assistive technology: Travel distance to obtain assistive products (%) - 26-50 km"]:
        return "mid_distance"
    
    if x in ["Assistive technology: Travel distance to obtain assistive products (%) - <6 km",
             "Assistive technology: Travel distance to obtain assistive products (%) - 6-25 km"]:
        return "close"

    if "Assistive technology: Need of assistive products (%)" in x:
        return "need"

    if "Assistive technology: Use of assistive products (%)" in x:
        return "use"

    if "Assistive technology: Met need of assistive products (%)" in x:
        return "met_need"
    
health_inequality["assistive_techno"] = health_inequality["indicator_name"].apply(assistive_techno)
```

### Simplification des catégories sur le diabète

```python
health_inequality.loc[
    health_inequality["indicator_name"].str.contains("Diabetes", na=False),
    "indicator_name"
].unique()
```

```python
def diabetes_simplification(x):
    if pd.isna(x):
        return np.nan

    x = x.lower()

    if "prevalence" in x:
        return "prevalence"

    if "treatment coverage" in x:
        return "treatment_coverage"

    return np.nan

health_inequality["diabetes"] = health_inequality["indicator_name"].apply(diabetes_simplification)
```

### Simplification des catégories sur l'hypertension

```python
health_inequality.loc[
    health_inequality["indicator_name"].str.contains("Hypertension", na=False),
    "indicator_name"
].unique()
```

```python
def hypertension_simplification(x):
    if pd.isna(x):
        return np.nan

    x = x.lower()

    if "prevalence" in x:
        return "prevalence"
    
    if "diagnosis coverage" in x:
        return "diagnosis"

    if "treatment coverage" in x and "effective" not in x:
        return "treatment"

    if "effective treatment" in x:
        return "effective_treatment"

    return np.nan
health_inequality["hypertension"] = health_inequality["indicator_name"].apply(hypertension_simplification)
```

### Simplification des catégories sur l'alcoolémie

Listons toutes les catégories existantes :

```python
health_inequality.loc[
    health_inequality["indicator_name"].str.contains("Alcohol", na=False),
    "indicator_name"
].unique()
```

```python
def alcohol_simplification(x):
    if pd.isna(x):
        return np.nan

    x = x.lower()

    if "abstainers" in x or "non-drinker" in x:
        return "no_or_light"

    if "consumers" in x or "average daily intake" in x:
        return "moderate"

    if "heavy episodic" in x or "heavy continuous" in x:
        return "heavy"

    if "harmful" in x or "attributable" in x or "cirrhosis" in x or "road traffic" in x or "dalys" in x:
        return "alcoholism"

    return np.nan

health_inequality["alcohol"] = health_inequality["indicator_name"].apply(alcohol_simplification)
```

### Simplification des catégories sur le tabaggisme

Listons toutes les catégories existantes :

```python
health_inequality.loc[
    health_inequality["indicator_name"].str.contains("Tobacco", na=False),
    "indicator_name"
].unique()
```

```python
def tobacco_simplification(x):
    if pd.isna(x):
        return np.nan

    x = x.lower()

    if "current tobacco use" in x and "daily" not in x and "smoking" not in x:
        return "non_smoker"

    if "current cigarette smoking" in x or "current smoking" in x:
        return "light_smoker"

    if "daily cigarette" in x or "daily tobacco smoking" in x:
        return "moderate_smoker"

    if "heavy" in x:
        return "heavy_smoker"

    return np.nan

health_inequality["tobacco"] = health_inequality["indicator_name"].apply(tobacco_simplification)
```
